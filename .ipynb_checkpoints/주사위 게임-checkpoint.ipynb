```python
import random
import numpy as np
import time

def dice_game(n):
    
    start = time.time()
    dice_result = new_input
    all_price = []        
    print("dice_result :", dice_result, "\n")
    
    for i in dice_result :
        
        before_get_a_result = []     
        for j in i:

            count = 0
            for k in i:

                if j == k:

                    count = count + 1

            if count >= 1:

                before_get_a_result.append((j, count))

        before_get_a_result = list(set(before_get_a_result))
        if len(before_get_a_result) == 3:

            three_number = []
            for i in before_get_a_result:

                three_number.append(i[0])

            max_number_in_three_number = max(three_number)
            all_price.append(max_number_in_three_number*100)

        elif len(before_get_a_result) == 2:

            for i in before_get_a_result:

                if i[1] == 2:

                    all_price.append(1000+(i[0]*100))
                    break

        elif len(before_get_a_result) == 1:

            for i in before_get_a_result:

                if i[1] == 3:

                    all_price.append(10000+(i[0]*1000))
                    break

    print("all_price :", all_price, "\n")
    print(time.time() - start)
    return max(all_price)
```

```python
# 해설 답안(map함수의 활용으로 최적화할 수 있음)
def dice_game2(n):
    
    start = time.time()
    dice_result = new_input
    all_price = []        
    print("dice_result :", dice_result, "\n")
    
    for i in dice_result :
        
        a, b, c = map(int, i)
        if a == b and b == c :
            
            all_price.append(10000+(i[0]*1000))
        
        elif a == b :
            
            all_price.append(1000+(a*100))
            
        elif a == c :
            
            all_price.append(1000+(a*100))
            
        elif b == c :
            
            all_price.append(1000+(b*100))
        
        else:
            
            three_dice_number_list = [a, b, c]
            max_number = max(three_dice_number_list)
            all_price.append(max_number * 100)
            
    max_price = max(all_price)
    print(time.time() - start)
    
    return max_price
```
