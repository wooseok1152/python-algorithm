```python
import sys

sys.stdin=open("input.txt", "r")
n = int(input())
a = list(map(int, input().split()))
m = int(input())
b = list(map(int, input().split()))

def sum_list(a, b):
    
    test_list1 = a
    test_list2 = b
    print("test_list1 :", test_list1, "\n")
    print("test_list2 :",test_list2, "\n")
    
    for i in range(len(test_list2)):
        
        test_list1.append(test_list2.pop())
        
    result = sorted(test_list1)
    for i in result:
        
        print(i, end = " ")
    
sum_list(a, b)
```
