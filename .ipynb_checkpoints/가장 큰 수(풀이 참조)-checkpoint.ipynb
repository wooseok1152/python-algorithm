```python
'''
input을 list로 변환하고 시작해야한다.
input_list의 0번째 index의 값을 pop하고 이를 stack에 append하는 것으로써, stack을 초기화한다.
반복문을 사용하여 input_list의 0번째를 pop하고, 이를 stack의 맨 마지막 index에 위치해 있는 값과 비교한다.
input_list에서 pop된 값보다 stack의 맨 마지막 index값이 작다면, stack의 맨 마지막 index의 값을 pop한다. 그리고 count를 1증가 시킨다. 이 과정을 while 반복문으로 실시한다.
input_list에서 pop된 값보다 stack의 맨 마지막 index값이 크다면, input_list에서 pop된 값을 해당 stack에 append한다.
제거 개수 조건에 부합하였는데, input_list에 요소가 남아있는 경우와 제거 개수 조건에 아직 부합하지 않은 경우도 고려해야 한다.
'''
import sys

sys.stdin = open(r"C:\Users\my\Desktop\Documents\파이썬 알고리즘 문제풀이(코딩테스트 대비)\섹션 5\1. 가장 큰 수\in1.txt")

input_int, m = input().split()
m = int(m)
input_list = []
for i in range(len(input_int)):
    
    input_list.append(int(input_int[i]))

def find_maximum(m, input_list):
    
    count = 0
    stack = []
    picked_number = input_list.pop(0)
    stack.append(picked_number)
    for i in range(0, len(input_list)):
        
        picked_number = input_list.pop(0)
        while True:
            
            if len(stack) == 0:
                
                stack.append(picked_number)
                break
                
            elif stack[-1] < picked_number:
                
                stack.pop()
                count = count + 1
                
            elif stack[-1] >= picked_number:
                
                stack.append(picked_number)
                break
            
            if count == m :
                
                stack.append(picked_number)
                break
        
        if count == m :
            
            break
    
    # 제거 개수 조건에 아직 부합하지 않은 경우
    if count != m:
        
        for i in range(m - count):
            
            stack.pop()
    
    # 제거 개수 조건에 부합하였는데, input_list에 요소가 남아있는 경우    
    if len(input_list) != 0:
        
        for i in input_list:
            
            stack.append(i)
        
    result = ""
    for i in stack:
        
        result = result + str(i)
        
    print(result)
    
find_maximum(m, input_list)
```

```python
'''강의 풀이'''

import sys

sys.stdin = open(r"C:\Users\my\Desktop\Documents\파이썬 알고리즘 문제풀이(코딩테스트 대비)\섹션 5\1. 가장 큰 수\in1.txt")

input_int, m = input().split()
m = int(m)
input_list = []
for i in range(len(input_int)):
    
    input_list.append(int(input_int[i]))
    
count = 0
stack = []

for i in range(len(input_list)):
    
    '''input_list의 맨 앞 요소를 pop한다.'''
    picked_value = input_list.pop(0)
    '''
    pop된 요소가 stack의 맨 위에 있는 요소보다 크다면 ,stack에 맨 위에 있는 요소를 pop한다. 
    그리고 이 과정을 'count == m'이 될 때까지 무한번 반복한다.
    '''
    while stack and stack[-1] < picked_value :
        
        stack.pop()
        count = count + 1
        if count == m:
            
            break
    
    stack.append(picked_value)
    if count == m:
        
        break
        
    if len(input_list) == 0:
        
        break
        
# 제거 개수 조건에 부합하였는데, input_list에 요소가 남아있는 경우          
if len(input_list) != 0:
    
    for i in input_list:
        
        stack.append(i)

# 제거 개수 조건에 아직 부합하지 않은 경우
if count != m :
    
    for i in range(m - count):
        
        stack.pop()
        
print(stack)
```
