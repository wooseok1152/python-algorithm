```python
def a_regular_polyhedron(n, m):
    
    # n 정다면체 숫자 리스트와 m 정다면체 숫자 리스트 생성
    n_number_list = []
    m_number_list = []
    for i in range(1, n+1):
        
        n_number_list.append(i)
        
    for j in range(1, m+1):
        
        m_number_list.append(j)
    
    print("n_number_list :", n_number_list, "\n")
    print("m_number_list :", m_number_list, "\n")
    
    # n 정다면체와 m 정다면체의 나올수 있는 모든 눈의 합을 요소로 가진 리스트 생성
    n_list_m_list_sum = []
    
    for i in n_number_list:
        
        for j in m_number_list:
            
            n_list_m_list_sum.append(i + j)
            
    print("n_list_m_list_sum :", n_list_m_list_sum, "\n")
    
    # 눈의 합 결과와 이에 대한 빈도수를 튜플로 묶고, 해당 튜플을 요소로 가진 리스트 생성
    frequency_for_each_sum_list = []
    for i in n_list_m_list_sum:
        
        count = 0
        for j in n_list_m_list_sum:
            
            if i == j:
                
                count = count + 1
                frequency_for_each_sum_list.append((i, count))
                
    print("frequency_for_each_sum_list :", frequency_for_each_sum_list,"\n")
    
    new_frequency_for_each_sum_list = []
    stack = []
    for i in range(2, n+m+1):
        
        for j in frequency_for_each_sum_list:
            
            if i == j[0]:
                
                stack.append(j)
        
        new_frequency_for_each_sum_list.append(stack[-1])
    
    print("new_frequency_for_each_sum_list :", new_frequency_for_each_sum_list, "\n")
    
    # 가장 높은 빈도수 구하기
    frequency = []
    for i in new_frequency_for_each_sum_list:
        
        frequency.append(i[1])
    
    max_frequency = max(frequency)
    
    _result = []
    
    # 가장 높은 빈도수를 가진 합 결과를 return하기
    for i in new_frequency_for_each_sum_list:
        
        if i[1] == max_frequency:
            
            _result.append(i)
            
    result = []
    for i in _result:
        
        result.append(i[0])
        
    return result
    
```

```python
a_regular_polyhedron(8, 12)
```
